#Lab:2 Plotting and Inferring emperical distribution



```python
#Emperical Distribution for Numerical data

import numpy as np
num_data = np.array([20,22,25,20,38,40,22,29,27,26,24,33,31])
```

```python
#Histogram
#Shows the frequency of values grouped into bins; approximates the emperical PDF
import matplotlib.pyplot as plt

plt.hist(num_data, bins=5, density=True)
plt.title("Histogram of Numerical Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
#Emperical CDF
#Shows the cumulative distribution of values; approximates the emperical CDF
x = np.sort(num_data)
y = np.arange(1, len(num_data)+1) / len(num_data)

plt.plot(x,y)
plt.title("ECDF - Numerical Data")
plt.xlabel("Value")
plt.ylabel("Cummulative Probability")
plt.show()
```

```python
from scipy.stats import gaussian_kde

kde= gaussian_kde(num_data)
x_values = np.linspace(min(num_data), max(num_data), 100)
y_values = kde(x_values)

plt.plot(x_values, y_values)
plt.title("Kernel Density Estimation (KDE)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
#Box Plot
#Shows median,quartiles and outliers

plt.boxplot(num_data)
plt.title("Box Plot of Numerical Data")
plt.ylabel("Value")
plt.show()
```

```python
#Violin plot

plt.violinplot(num_data)
plt.title("Violin Plot of Numerical Data")
plt.ylabel("Value")
plt.show()
```
